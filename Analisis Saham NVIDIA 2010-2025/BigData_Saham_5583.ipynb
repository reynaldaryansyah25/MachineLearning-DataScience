# Penjelasan tentang data set


Ini adalah Data Pasar Saham Historis dari lima perusahaan Teknologi Besar: NVIDIA (NVDA), Apple (AAPL), Microsoft (MSFT), Google (GOOGL), dan Amazon (AMZN) selama 15 tahun dari 1 Januari 2010 hingga 1 Januari 2025.
Termasuk data saham harian dengan harga pembukaan dan penutupan, harga tertinggi, terendah, dan volume perdagangan.
Kumpulan data ini berfungsi sebagai sumber daya yang berharga untuk menganalisis tren pertumbuhan jangka panjang, volatilitas, dan perilaku pasar raksasa teknologi terkemuka.
Dengan menganalisis kumpulan data ini, kita dapat memperoleh pemahaman lebih mendalam tentang perilaku saham historis NVDA , AAPL , MSFT , GOOGL , dan AMZN selama 15 tahun dan membuat prediksi tentang kinerja masa depan mereka.

Date: Tanggal perdagangan.

Open_X: Harga saham saat pasar dibuka pada hari itu.

High_X: Harga tertinggi saham yang dicapai hari itu.

Low_X: Harga terendah saham yang dicapai hari itu.

Close_X: Harga saham saat penutupan pasar.

Volume_X: Jumlah saham yang diperdagangkan hari itu.

## Outlier

```python
# Kolom harga penutupan
close_columns = [f'Close_{ticker}' for ticker in ['AAPL', 'AMZN', 'GOOGL', 'MSFT', 'NVDA']]

# Hitung return harian
returns = df[close_columns].pct_change().dropna()

# Ubah ke format long
returns_melted = returns.melt(var_name='Saham', value_name='Return')
returns_melted['Saham'] = returns_melted['Saham'].str.replace('Close_', '')

Q1 = df['Close_NVDA'].quantile(0.25)
Q3 = df['Close_NVDA'].quantile(0.75)
IQR = Q3 - Q1

df_cleaned = df[(df['Close_NVDA'] >= Q1 - 1.5 * IQR) & (df['Close_NVDA'] <= Q3 + 1.5 * IQR)]


# Visualisasi boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=returns_melted, x='Saham', y='Return', palette='Set2')
plt.title('Boxplot Return Harian Saham')
plt.ylabel('Return Harian')
plt.xlabel('Saham')
plt.tight_layout()
plt.show()
```

```python
# Moving Averages
df['MA50'] = df['Close_NVDA'].rolling(window=10).mean()
df['MA200'] = df['Close_NVDA'].rolling(window=100).mean()

# Visualisasi Moving Averages
plt.figure(figsize=(14,6))
plt.plot(df['Date'], df['Close_NVDA'], label='Harga Penutupan', color='blue')
plt.plot(df['Date'], df['MA50'], label='MA 50 Hari', color='orange')
plt.plot(df['Date'], df['MA200'], label='MA 200 Hari', color='green')
plt.title('Harga NVDA dengan MA50 & MA200')
plt.xlabel('Tanggal')
plt.ylabel('Harga ($)')
plt.legend()
plt.grid(True)
plt.show()


```

```python
# RSI
delta = df['Close_NVDA'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
rs = gain / loss
df['RSI'] = 100 - (100 / (1 + rs))

# Visualisasi RSI
plt.figure(figsize=(14,4))
plt.plot(df['Date'], df['RSI'], color='purple', label='RSI')
plt.axhline(70, color='red', linestyle='--', label='Overbought')
plt.axhline(30, color='green', linestyle='--', label='Oversold')
plt.title('RSI NVIDIA')
plt.xlabel('Tanggal')
plt.ylabel('RSI')
plt.legend()
plt.grid(True)
plt.show()
```

# Membangun model multipled linear regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
features = ['Open_NVDA', 'High_NVDA', 'Low_NVDA']
target = 'Close_NVDA'

print("Features used for prediction:", features)
print("Target variable:", target)
```

```python
X = df[features]
y = df[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
print('Coefisient: ', lr_model.coef_)
print('Intercept: ', lr_model.intercept_)
```

# Perbandingan data prediksi dan actual

```python
dates_test = df.loc[y_test.index, 'Date']

y_pred = lr_model.predict(X_test)

print("Data Actual/Asli Variabel harga: \n", y_test[0:5])
print("\n")
print("Data Prediksi Variabel harga: \n", y_pred[0:5])

plt.figure(figsize=(12, 5))
plt.plot(dates_test, y_test, label='Actual', linewidth=2)
plt.plot(dates_test, y_pred, label='Predicted', linestyle='--', linewidth=2)
plt.title('Memprediksi NVDIA dengan Fitur Historis')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()

# Format tanggal pada sumbu x
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Evaluasi Model

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score,mean_absolute_percentage_error

print('Root Mean Squared Error (RMSE) :\t %.4f' % np.sqrt(mean_squared_error(y_pred, y_test)))
print('Mean Absolute Error (MAE)\t\t : %.4f' % mean_absolute_error(y_pred, y_test))
print('Mean Squared Error (MSE)\t\t : %.4f' % mean_squared_error(y_pred, y_test))
print('Mean Absolute Percentage Error (MAPE)\t : %.4f' % mean_absolute_percentage_error (y_test, y_pred))
print('R Square (Koefisien Determinasi)\t : %.4f' % r2_score(y_test, y_pred))
```

```python
# Evaluasi Error (Residual)
error = y_test - y_pred

plt.figure(figsize=(10,5))
plt.plot(y_test.index, error, marker='o', linestyle='-', color='red')
plt.axhline(y=0, color='black', linestyle='--')
plt.title('Plot Error Residual: Aktual - Prediksi')
plt.xlabel('Index')
plt.ylabel('Error')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.hist(error, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribusi Error Residual')
plt.xlabel('Error')
plt.ylabel('Frekuensi')
plt.grid(True)
plt.show()
```
